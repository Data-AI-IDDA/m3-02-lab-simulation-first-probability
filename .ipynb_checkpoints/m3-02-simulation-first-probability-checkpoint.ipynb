task1

```python
import numpy as np
import matplotlib.pyplot as plt

N = 100_000

x = np.random.uniform(-1, 1, N)
y = np.random.uniform(-1, 1, N)

inside = (x**2 + y**2) <= 1

points_inside = np.sum(inside)
pi_estimate = 4 * points_inside / N

print(f"Estimated π: {pi_estimate}")
print(f"True π: {np.pi}")
```

```python
plt.figure(figsize=(6,6))

n_plot = 5000

inside_plot = inside[:n_plot]
outside_plot = ~inside_plot

plt.scatter(x[:n_plot][inside_plot], y[:n_plot][inside_plot], s=5, label="Inside")
plt.scatter(x[:n_plot][outside_plot], y[:n_plot][outside_plot], s=5, label="Outside")

theta = np.linspace(0, 2*np.pi, 200)
circle_x = np.cos(theta)
circle_y = np.sin(theta)
plt.plot(circle_x, circle_y, linewidth=2)

plt.title("Monte Carlo π Estimation (First 5000 Points)")
plt.legend()
plt.axis("equal")
plt.show()
```

```python
running_pi = []
steps = []

count_inside = 0

for i in range(N):
    if inside[i]:
        count_inside += 1
    
    if (i + 1) % 100 == 0:
        pi_val = 4 * count_inside / (i + 1)
        running_pi.append(pi_val)
        steps.append(i + 1)

plt.figure(figsize=(8,5))
plt.plot(steps, running_pi, label="Estimated π")
plt.axhline(y=np.pi, linestyle="--", label="True π")

plt.xlabel("Number of Points")
plt.ylabel("π Estimate")
plt.title("Convergence of π Estimate")
plt.legend()
plt.show()
```

Guiding question:
The estimate stabilizes within ±0.01 of the true value of π at roughly 5,000–8,000 samples.

task2

```python
N = 10000

flips = np.random.binomial(1, 0.5, size=N)
cumulative_heads = np.cumsum(flips)
flip_counts = np.arange(1, N + 1)
running_proportion = cumulative_heads / flip_counts

plt.figure(figsize=(8,5))

plt.plot(flip_counts, running_proportion, label="Running Proportion")

plt.axhline(y=0.5, linestyle="--", label="True Probability (0.5)")

plt.xscale("log")

plt.xlabel("Number of Flips (log scale)")
plt.ylabel("Proportion of Heads")
plt.title("Convergence of Coin Flip Probability")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5))

for i in range(5):
    flips = np.random.binomial(1, 0.5, size=N)
    
    cumulative_heads = np.cumsum(flips)
    flip_counts = np.arange(1, N + 1)
    running_proportion = cumulative_heads / flip_counts
    
    plt.plot(flip_counts, running_proportion, label=f"Run {i+1}")

plt.axhline(y=0.5, linestyle="--", label="True Probability (0.5)")
plt.xscale("log")

plt.xlabel("Number of Flips (log scale)")
plt.ylabel("Proportion of Heads")
plt.title("Convergence of Coin Flip Probability (5 Runs)")
plt.legend()
plt.show()
```

From the convergence plot, the running proportion typically stays within ±0.01 of the true probability (0.5) after roughly 2,000–3,000 coin flips.

task3

```python
from scipy.stats import binom

n_widgets = 50       
p_defect = 0.03      
n_batches = 10000 

defect_counts = np.random.binomial(n=n_widgets, p=p_defect, size=n_batches)


plt.figure(figsize=(8,5))
counts, bins, patches = plt.hist(defect_counts, bins=np.arange(-0.5, max(defect_counts)+1.5, 1),
                                 density=True, alpha=0.6, color='skyblue', edgecolor='black', label='Simulation')

k = np.arange(0, max(defect_counts)+1)
pmf = binom.pmf(k, n_widgets, p_defect)
plt.plot(k, pmf, 'o-', color='red', label='Theoretical Binomial PMF')

plt.xlabel("Number of Defects per Batch")
plt.ylabel("Probability / Frequency")
plt.title("Distribution of Defects in 10,000 Batches")
plt.legend()
plt.show()
```

```python
P0_empirical = np.mean(defect_counts == 0)
P3plus_empirical = np.mean(defect_counts >= 3)
expected_defects_empirical = np.mean(defect_counts)

P0_theoretical = binom.pmf(0, n_widgets, p_defect)
P3plus_theoretical = 1 - binom.cdf(2, n_widgets, p_defect)
expected_defects_theoretical = n_widgets * p_defect

print(f"P(0 defects) → Empirical: {P0_empirical:.4f}, Theoretical: {P0_theoretical:.4f}")
print(f"P(≥3 defects) → Empirical: {P3plus_empirical:.4f}, Theoretical: {P3plus_theoretical:.4f}")
print(f"Expected defects per batch → Empirical: {expected_defects_empirical:.2f}, Theoretical: {expected_defects_theoretical:.2f}")
```

The empirical estimates are very close to the theoretical values from the Binomial distribution. The differences are very small (around 0.003–0.01), showing that the simulation matches the theoretical model well. While 1,000 batches could give a reasonable approximation, using 10,000 batches provides more stable and accurate estimates due to reduced random variation.
