### TASK1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
sns.set_style("whitegrid")

N = 100000

# random points
x = np.random.uniform(-1, 1, N)
y = np.random.uniform(-1, 1, N)

# inside circle
inside = x**2 + y**2 <= 1

# estimate pi
pi_estimate = 4 * np.sum(inside) / N

print("Estimated π:", pi_estimate)
print("True π:", np.pi)

plt.figure(figsize=(6,6))

plt.scatter(x[:5000][inside[:5000]], y[:5000][inside[:5000]], s=1)
plt.scatter(x[:5000][~inside[:5000]], y[:5000][~inside[:5000]], s=1)

# circle
theta = np.linspace(0, 2*np.pi, 100)
plt.plot(np.cos(theta), np.sin(theta))

plt.title("Monte Carlo π estimation")
plt.show()

```

```python
running_pi = []
counts = []

for i in range(100, N+1, 100):
    pi_i = 4 * np.sum(inside[:i]) / i
    running_pi.append(pi_i)
    counts.append(i)

plt.plot(counts, running_pi)
plt.axhline(np.pi)
plt.ylim(3.10,3.16)
plt.xlabel("Number of samples")
plt.ylabel("Estimated π")
plt.title("Convergence of π estimate")
plt.show()

```

### Guiding Question Answer

By observing the convergence plot, the estimate of π fluctuates at the beginning due to small sample size. As more points are added, the running estimate stabilizes closer to the true value of π.

Rough stabilization point: Around 30,000–40,000 samples, the running estimate generally stays within ±0.01 of π.

Observation: Before ~30,000 samples, fluctuations can be larger than ±0.01. After that, the estimate converges and remains consistently close to the true π.

### TASK 2

```python
flips = np.random.binomial(1, 0.5, size=10000)

running_heads = np.cumsum(flips) / np.arange(1, 10001)

plt.plot(running_heads)
plt.axhline(0.5)

plt.xscale("log")
plt.title("Running probability of heads")
plt.show()
```

```python
for _ in range(5):
    flips = np.random.binomial(1, 0.5, size=10000)
    running_heads = np.cumsum(flips) / np.arange(1, 10001)
    plt.plot(running_heads)

plt.axhline(0.5)
plt.xscale("log")
plt.title("5 simulations")
plt.show()
```

### Guiding Question Answer

By examining the running proportion plots for coin flips:

Initial fluctuations are large due to the small number of flips.

As more flips are performed, the empirical proportion of heads approaches the theoretical probability 0.5.

Typical stabilization point: Around 4,000–5,000 flips, the running proportion generally stays within ±0.01 of 0.5.

Observation: Before ~4,000 flips, the running proportion may fluctuate outside the ±0.01 range, but after that, it remains relatively stable.

### TASK 3

```python
batches = 10000
data = np.random.binomial(50, 0.03, size=batches)

sns.histplot(data, bins=20, stat="probability")
plt.title("Defect distribution")
plt.show()

p0 = np.mean(data == 0)
p3_plus = np.mean(data >= 3)
expected = np.mean(data)

print("P(0 defects):", p0)
print("P(3+ defects):", p3_plus)
print("Expected defects:", expected)
```

```python
from scipy.stats import binom

p0_theory = binom.pmf(0, 50, 0.03)
p3_plus_theory = 1 - binom.cdf(2, 50, 0.03)
expected_theory = 50 * 0.03

print(p0_theory, p3_plus_theory, expected_theory)
```
